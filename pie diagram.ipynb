```python
from matplotlib import pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
```

```python
slices = [40,50,10]
labels = ['forty','fifty','ten']
colors = ['#45B39D','#5DADE2','#EC7063']

plt.pie(slices, 
        labels = labels, 
        colors = colors,
        wedgeprops = {'edgecolor':'black'})

plt.title('Pie diagram')
plt.tight_layout()
plt.show()
```

### explode a patricular piece and add shadow and add angle

```python
slices = [40,50,10]
labels = ['forty','fifty','ten']
colors = ['#45B39D','#5DADE2','#EC7063']

#explode
explode = [0,0,0.2]

plt.pie(slices, 
        labels = labels, 
        colors = colors,
        
        #new
        explode = explode,
        shadow = True,
        startangle = 90,
        
        wedgeprops = {'edgecolor':'black'})

plt.title('Pie diagram')
plt.tight_layout()
plt.show()
```

### percentage of pie 

```python
slices = [40,50,10]
labels = ['forty','fifty','ten']
colors = ['#45B39D','#5DADE2','#EC7063']

explode = [0,0,0.2]

plt.pie(slices, 
        labels = labels, 
        colors = colors,
        explode = explode,
        shadow = True,
        startangle = 90,
        
        #percentage #syntax is strange but use copy paste 
        autopct = '%1.1f%%',
        
        wedgeprops = {'edgecolor':'black'})

plt.title('Pie diagram')
plt.tight_layout()
plt.show()
```
