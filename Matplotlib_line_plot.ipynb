### import libraries

```python
from matplotlib import pyplot as plt
```

### Plot with label and Title

```python
x = list(range(1,10))
y = list(range(11,20))

# LABELLING 
plt.xlabel('x-axis')
plt.ylabel('y-axis')

#TITLE
plt.title('Title of Plot')

#PLOTTING
plt.plot(x,y)
# use ; to elimintate the text above plot
# plt.plot(x,y);
```

### Multiple line plot in single plot

```python
x = list(range(1,10))
y1 = list(range(11,20))
y2 = list(range(21,30))

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Title of Plot')

plt.plot(x,y1, label = 'x,y1 line') #add the legend label here only so that no confusion will arise
plt.plot(x,y2, label = 'x,y2 line')

plt.legend();

# alternative 
# plt.legend(['x,y1 line','x,y2 line']);
```

### Format Strings - [marker][line][color] // grid // padding

```python
x = list(range(1,10))
y1 = list(range(11,20))
y2 = list(range(21,30))

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Title of Plot')

# changed color [can use hex codes also] / linestyle / marker / linewidth
# check string format in this link https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html
plt.plot(x,y1, color='r',linestyle = '--',marker='o' , linewidth = 3, label = 'x,y1 line') 
plt.plot(x,y2,color='#444444',linestyle = '-.',marker='s' , label = 'x,y2 line')

#padding 
plt.tight_layout()
#grid lines
plt.grid(True)

plt.legend();
```

### using built in style for plots and saving figure

```python
x = list(range(1,10))
y1 = list(range(11,20))
y2 = list(range(21,30))

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Title of Plot')

print(plt.style.available)
plt.style.use('seaborn-darkgrid')

plt.plot(x,y1, color='r',linestyle = '--',marker='o' , linewidth = 3, label = 'x,y1 line') 
plt.plot(x,y2,color='#444444',linestyle = '-.',marker='s' , label = 'x,y2 line')

#grid lines
plt.grid(True)
plt.legend();

# saving fig
# plt.savefig('fig.png')

plt.show()
```

### stack plots

```python
import numpy as np
from matplotlib import pyplot as plt

minutes = np.arange(6)

p1 = [1,2,3,3,4,5]
p2 = [1,2,3,4,4,5]
p3 = [1,2,3,3,5,5]

labels = ['p1','p2','p3']
colors = ['#CCCCFF','#DFFF00','#9FE2BF']

plt.stackplot(minutes,p1,p2,p3, labels= labels, colors = colors)

plt.title('Stack plot')
plt.tight_layout()

# change location of legend with loc command
plt.legend(loc='upper left')
# location can also be specified ---
#plt.legend(loc=(0.07,0.05))

plt.show()
```

### filling area on line plots

```python
x = list(range(1,10))
y1 = list(range(11,20))
y2 = list(range(21,30))

plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Title of Plot')

print(plt.style.available)
plt.style.use('seaborn-darkgrid')

plt.plot(x,y1, 
         color='r',
         linestyle = '--',
         marker='o', 
         linewidth = 3, 
         label = 'x,y1 line') 
plt.plot(x,y2,
         color='#444444',
         linestyle = '-.',
         marker='s' ,
         label = 'x,y2 line')

#filling area b.w line plots
# alpha = 0.25 gives transparency 
plt.fill_between(x,y1, color = 'r', alpha=0.25)
# where=(condition), interpolate = True >> will fill area only for required 

plt.grid(True)
plt.legend();
plt.show()
```
