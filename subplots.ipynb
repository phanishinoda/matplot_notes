```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
x = list(range(1,5))
y1 = list(range(6,10))
y2 = list(range(11,15))
```

### share y and horizontal plots

```python
plt.style.use('seaborn')

#define grid
fig,(ax1,ax2) = plt.subplots(nrows=1,ncols=2, sharey = True)

#plot for axis1
ax1.plot(x,y1, label="x,y1")
ax1.legend()
ax1.set_title("title")

#plot for axis2
ax2.plot(x,y2, label="x,y2")
ax2.legend()
ax2.set_title("title")

plt.tight_layout()
plt.show()
```

### sharex with vertical plots

```python
plt.style.use('seaborn')

#define grid number of rows and cols 
fig,(ax1,ax2) = plt.subplots(nrows=2,ncols=1, sharex = True)

#plot for axis1
ax1.plot(x,y1, label="x,y1")
ax1.legend()
ax1.set_title("title")

#plot for axis2
ax2.plot(x,y2, label="x,y2")
ax2.legend()

plt.tight_layout()
plt.show()
```
