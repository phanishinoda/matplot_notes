### bar plot with basic labels

```python
x = list(range(1,10))
y1 = list(range(11,20))
y2 = list(range(21,30))

#change plot to bar 
plt.bar(x,y1, color = '#444444', label = 'x,y1 bar')


# axis-labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Title')
plt.show()
```

### side by side bars

```python
import numpy as np

x = list(range(1,10))
y1 = list(range(11,20))
y2 = list(range(21,30))
y3 = list(range(31,40))

# create a array of number for x axis
x_labels = np.arange(len(x))
width = 0.25

# change x to x_labels and width of the bar
plt.bar(x_labels-width,y1,  width = width,
        color = '#444444', 
        label = 'x,y1 bar') # to the left 
plt.bar(x_labels,y2,  width = width,
        color = '#76D7C4', 
        label = 'x,y2 bar') #center
plt.bar(x_labels+width,y3, width = width,
        color = '#F5B041', 
        label = 'x,y2 bar') # to the right

# axis-labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Title')
plt.show()
```

### Stacked Bar chart

#### 2columns data

```python
x = ['A', 'B', 'C', 'D']
y1 = [10, 20, 10, 30]
y2 = [20, 25, 15, 25]
  
# plot bars in stack manner
plt.bar(x, y1, 
        color='#F4D03F')
plt.bar(x, y2, 
        bottom=y1, 
        color='#CACFD2')
plt.show()
```

#### 3 columns data

```python
import numpy as np

x = ['A', 'B', 'C', 'D']
y1 = np.array([10, 20, 10, 30])
y2 = np.array([20, 25, 15, 25])
y3 = np.array([12, 15, 19, 6])
y4 = np.array([10, 29, 13, 19])
  
# plot bars in stack manner
plt.bar(x, y1, color='#444444')
plt.bar(x, y2, bottom=y1, color='#F4D03F')
plt.bar(x, y3, bottom=y1+y2, color='#CACFD2')
plt.bar(x, y4, bottom=y1+y2+y3, color='#3498DB')
plt.xlabel("Teams")
plt.ylabel("Score")
plt.title("Scores by Teams in 4 Rounds")
plt.show()
```

#### 3 columns with dataframe

```python
import pandas as pd
  
# dataframe
df = pd.DataFrame([['A', 10, 20, 10, 26], 
                   ['B', 20, 25, 15, 21], 
                   ['C', 12, 15, 19, 6],
                   ['D', 10, 18, 11, 19]],
                  columns=['Team', 'Round 1', 'Round 2', 'Round 3', 'Round 4'])
# view data
print(df)

colors = ['#D98880','#EC7063','#AF7AC5','#5DADE2']

# plot data in stack manner of bar type
df.plot(x='Team', 
        kind='bar', 
        stacked=True,
        color = colors,
        title='Stacked Bar Graph by dataframe')
```
